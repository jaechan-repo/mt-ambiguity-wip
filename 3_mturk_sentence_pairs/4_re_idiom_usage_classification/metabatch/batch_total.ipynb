```python
import pandas as pd
NUM_ASSIGNMENTS = 2

dfs = [pd.read_csv("../batch2/batch2_agg.csv"),
    pd.read_csv("../batch3/batch3_agg.csv"),
    pd.read_csv("../batch4/batch4+1_agg.csv")]

for i in range(3):
    dfs[i]['batch_num'] = i + 2

df = pd.concat(
    dfs,
    axis=0
)
df.sort_values(by=['idiom', 'label'], inplace=True)
df
```

##### Inter-Annotator Agreement Rate

```python
labels = ['figurative', 'literal', 'ambiguous', 'discard', 'tie']
golds = df[[f'gold_{i}' for i in range(NUM_ASSIGNMENTS)]].copy()
def g(row):
    for i in range(NUM_ASSIGNMENTS):
        row[f'gold_{i}'] = labels.index(row[f'gold_{i}'])
    return row
golds = golds.apply(g, axis=1)
golds.head()
```

```python
from statsmodels.stats.inter_rater import aggregate_raters
from statsmodels.stats.inter_rater import fleiss_kappa
golds = aggregate_raters(golds)[0]
kappa = fleiss_kappa(golds, method='fleiss')
kappa
```
