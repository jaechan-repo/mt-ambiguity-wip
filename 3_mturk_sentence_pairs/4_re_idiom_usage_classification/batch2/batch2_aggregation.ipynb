```python
import pandas as pd
import numpy as np

NUM_PROBLEMS = 5
NUM_ASSIGNMENTS = 2
```

```python
cols = ['idiom_0',
       'instance_0', 'label_0', 'meaning_0', 'idiom_1', 'instance_1',
       'label_1', 'meaning_1', 'idiom_2', 'instance_2', 'label_2', 'meaning_2',
       'idiom_3', 'instance_3', 'label_3', 'meaning_3', 'idiom_4',
       'instance_4', 'label_4', 'meaning_4', '0_gold', '1_gold',
        '2_gold', '3_gold', '4_gold',
       'TimeMe', 'AssignmentId', 'WorkerId']

column_groups = [
        ['idiom_0', 'instance_0', 'label_0', 'meaning_0', '0_gold'],
        ['idiom_1', 'instance_1', 'label_1', 'meaning_1', '1_gold'],
        ['idiom_2', 'instance_2', 'label_2', 'meaning_2', '2_gold'],
        ['idiom_3', 'instance_3', 'label_3', 'meaning_3', '3_gold'],
        ['idiom_4', 'instance_4', 'label_4', 'meaning_4', '4_gold'],
    ]

# common columns for all rows
common_cols = ['TimeMe', 'AssignmentId', 'WorkerId']
```

```python
def str_replace(df):
    df.columns = df.columns.str.replace('Input.', '') 
    df.columns = df.columns.str.replace('Answer.', '')
    return df

df = pd.read_csv("batch2_result.csv")
df = str_replace(df)
df = df[cols]
df = df.fillna('{}')
df.head()
```

```python
def melt(df):
    # list of column sets

    # melt the dataframe
    dfs = []
    for i, column_group in enumerate(column_groups):
        melted = df[column_group + common_cols].copy()  # select the columns
        melted.columns = ['idiom', 'instance', 'label', 'meaning', 'gold',]+ common_cols  # rename the columns
        melted['group'] = i  # add a column to indicate the original group
        dfs.append(melted)

    # concatenate the melted dataframes
    df_melted = pd.concat(dfs, ignore_index=True)
    return df_melted.drop('group', axis=1)

df = melt(df)
df.head()
```

```python
df = df.groupby(['idiom', 'instance', 'label', 'meaning']).agg(list).reset_index()
df['len'] = df['gold'].apply(len)
df['len'].value_counts()
```

```python
from collections import Counter

def most_frequent(array):
    count = Counter(array)
    max_count = max(count.values())
    max_elements = [k for k, v in count.items() if v == max_count]

    if len(max_elements) > 1:
        return 'tie'  # return -1 in case of a tie

    return max_elements[0]  # return the most frequent element
```

```python
labels = ['figurative', 'literal', 'ambiguous', 'discard']
def f(row):
    for i in range(NUM_ASSIGNMENTS):
        row[f'gold_{i}'] = labels[row['gold'][i]]

    row['label'] = labels[row['label']]
    row['gold'] = row['gold'][:NUM_ASSIGNMENTS]
    row['TimeMe'] = row['TimeMe'][:NUM_ASSIGNMENTS]
    row['AssignmentId'] = row['AssignmentId'][:NUM_ASSIGNMENTS]
    row['WorkerId'] = row['WorkerId'][:NUM_ASSIGNMENTS]
    row['maxVote'] = most_frequent([
        row[f'gold_{i}'] for i in range(NUM_ASSIGNMENTS)])
    return row
df = df.apply(f, axis=1).drop('len', axis=1)
df.head()
```

```python
golds = df[[f'gold_{i}' for i in range(NUM_ASSIGNMENTS)]].copy()
def g(row):
    for i in range(NUM_ASSIGNMENTS):
        row[f'gold_{i}'] = labels.index(row[f'gold_{i}'])
    return row
golds = golds.apply(g, axis=1)

from statsmodels.stats.inter_rater import aggregate_raters
from statsmodels.stats.inter_rater import fleiss_kappa
golds = aggregate_raters(golds)[0]
kappa = fleiss_kappa(golds, method='fleiss')
kappa
```
