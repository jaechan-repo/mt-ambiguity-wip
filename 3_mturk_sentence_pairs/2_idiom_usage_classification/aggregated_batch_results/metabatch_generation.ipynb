```python
import pandas as pd
batch1 = pd.read_csv("batch_1_summary/batch_1_instance.csv")
batch2 = pd.read_csv("batch_2_summary/batch_2_instance.csv")
batch3 = pd.read_csv("batch_3_summary/batch_3_instance.csv")
batch4 = pd.read_csv("batch_4_summary/batch_4_instance.csv")
batches = [batch1, batch2, batch3, batch4]
NUM_ASSIGNMENTS = 3
```

```python
metabatch = pd.concat([
    batch1, batch2, batch3, batch4
]).reset_index(drop=True)
metabatch = metabatch.sort_values(['idiom', 'label']).reset_index(drop=True)
metabatch.to_csv('metabatch_result.csv', index=False)
metabatch
```

```python
from statsmodels.stats.inter_rater import aggregate_raters
from statsmodels.stats.inter_rater import fleiss_kappa
golds = aggregate_raters(metabatch[[(f'gold_{i}') for i in range(NUM_ASSIGNMENTS)]])[0]
kappa = fleiss_kappa(golds, method='fleiss')
kappa
```
